```python
import matplotlib.pyplot as plt
import seaborn as sns
import scanpy as sc
import numpy as np
import pandas as pd
import surgeon
```

```python
sns.set(style="whitegrid")
```

# Freezed scNet scores

```python
scnet_freezed = pd.read_csv("./results/convergence/pancreas/scores_scNetFreezed_count.log")
scnet_freezed['Architecture'] = "Freezed scNet"
scnet_freezed
```

# UnFreezed scNet scores

```python
scnet_unfreezed = pd.read_csv("./results/convergence/toy/scores_scNetUnFreezed_count.log")
scnet_unfreezed['Architecture'] = "UnFreezed scNet"
scnet_unfreezed
```

# scVI scores

```python
scvi_unfreezed = pd.read_csv("./results/convergence/pancreas/scVI.csv")
scvi_unfreezed['Architecture'] = "scVI"
scvi_unfreezed['time'] = scvi_unfreezed['Elapsed Time']
scvi_unfreezed
```

# Integrate scores

```python
df = pd.concat([scnet_freezed,
                scnet_unfreezed,
                scvi_unfreezed])
df
```

# Convergence plot 

```python
g = sns.relplot(x="time", y="ASW",
            hue="Architecture", 
            palette=sns.color_palette("rocket_r", 3),
            height=8, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=df)
plt.xticks(np.arange(0, 60, 5), rotation=90)
# plt.yticks(np.arange(0, .9, 0.1))
plt.xlabel("time (seconds)")
g.despine(left=True)
```
